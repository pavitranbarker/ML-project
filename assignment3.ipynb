```python
import pandas as pd
df=pd.read_csv("healthcare_dataset.csv")
print(df)
```

```python
df.shape
```

```python
num_cols = ["Salary", "Age", "Income"]
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
num_cols=["Age","Income","Salary"]
df[num_cols]=df[num_cols].fillna(df[num_cols].median())
```

```python
cat_cols = ["Gender", "Job_Role", "City", "State"]
for col in num_cols:
  df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df['Gender']=df['Gender'].replace({
    "M":"Male",
    "male":"Male",
    "f":"Female",
    "female":"Female"
})
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Education_Level']=le.fit_transform(df['Education_Level'])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoding_col=["Education_Level","Department"]
le=LabelEncoder()
for col in encoding_col:
  df[col]=le.fit_transform(df[col])
df.head()
```

```python
#binary encoding
df['Loan_Status']=df['Loan_Status'].map({"Rejected":0,"Approved":1})
df.head()
```

```python
df['Marital_Status']=df['Marital_Status'].map({"Married":1,"Single":0})
df.head()
```

```python
df=pd.get_dummies(df,columns=['Blood_Group'],drop_first=True)
df.head()
```
